Line Chart

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(10,90,5)
# y = np.random.randint(10,50,2)

print(x)

plt.plot(x, linestyle = "dashed" , color = "black") #linestyle = "dotted/dashed"  #To plot the line graph

plt.show()  #To show the plotted graph
```

Bar Chart

```python
#Bar chart requires 1 categorical & 1 numerical data
x = np.array(["Dogs","Cats","Hamster","Fish","Rats"])
y = np.random.randint(10,90,5)
```

```python
plt.bar(x,y)
```

Scatter Plot

```python
x = np.random.randint(1,20,50)
y = np.random.randint(1,20,50)
print(x,y)
plt.scatter(x,y)
```

Pie Chart

```python
#One numerical is required for pie
p = np.random.randint(1,20,5)
plt.pie(p , labels = ["Steel" , 'Agriculture' , "FMCG" , "Finance" , "IT" ])
```

Histogram

```python
x = np.random.normal(17 , 170 ,230)
plt.hist(x)
```
