```python
python -m spacy download en_core_web_lg
```

```python
nlp = spacy.load("en_core_web_lg")
```

```python
doc = nlp("She folded her handkerchief neatly.")
for i in doc:
    print(i.text , "  |  ",i.has_vector,"  |  ",i.is_oov)
```

```python
doc = nlp("bread egg burger sandwich boy")
base = nlp("bread")
for i in doc:
    print(f"{i.text} <-> {base.text} = {i.similarity(base)}")
```
